### Проверка расчётов

Проверьте, что эффективность обогащения рассчитана правильно. Вычислите её на обучающей выборке для признака `rougher.output.recovery`. Найдите `MAE` между вашими расчётами и значением признака. Опишите выводы.

**Описание данных**  
**Технологический процесс**

    Rougher feed — исходное сырье
    Rougher additions (или reagent additions) — флотационные реагенты: Xanthate, Sulphate, Depressant
      
        Xanthate **— ксантогенат (промотер, или активатор флотации);
        Sulphate — сульфат (на данном производстве сульфид натрия);
        Depressant — депрессант (силикат натрия).
    Rougher process (англ. «грубый процесс») — флотация
    Rougher tails — отвальные хвосты
    Float banks — флотационная установка
    Cleaner process — очистка
    Rougher Au — черновой концентрат золота
    Final Au — финальный концентрат золота

**Параметры этапов**

    air amount — объём воздуха
    fluid levels — уровень жидкости
    feed size — размер гранул сырья
    feed rate — скорость подачи

**Наименование признаков**  
Наименование признаков должно быть такое:  
[этап].[тип_параметра].[название_параметра]  
Пример: rougher.input.feed_ag  
Возможные значения для блока [этап]:  

    rougher — флотация
    primary_cleaner — первичная очистка
    secondary_cleaner — вторичная очистка
    final — финальные характеристики

Возможные значения для блока [тип_параметра]:

    input — параметры сырья
    output — параметры продукта
    state — параметры, характеризующие текущее состояние этапа
    calculation — расчётные характеристики

```python
# для расчёта эффективности Recovery воспользуюсь формулойЮ сразу записав её в функцию

def recovery(C, F, T):
    recovery_out = ((C*(F - T))/(F*(C - T))*100)
    
    return recovery_out
```

где:

    C — доля золота в концентрате после флотации/очистки;
    F — доля золота в сырье/концентрате до флотации/очистки;
    T — доля золота в отвальных хвостах после флотации/очистки.

Для прогноза коэффициента нужно найти долю золота в концентратах и хвостах. Причём важен не только финальный продукт, но и черновой концентрат.

```python
#выбираю данные и заношу из в отдельный датафрейм
data_train_test = data_train[['rougher.output.concentrate_au',
                              'rougher.input.feed_au',
                              'rougher.output.tail_au',
                              'rougher.output.recovery']]
# Считаю recovery
data_train_test['check_recovery'] = data_train_test.apply(
    lambda row: recovery(row['rougher.output.concentrate_au'], 
                                row['rougher.input.feed_au'], 
                                row['rougher.output.tail_au']), 
    axis=1)

data_train_test.head()
```

```python
# делю на целевой и предсказанные выборки
data_test_target = data_train_test ['rougher.output.recovery']
data_test_predicted = data_train_test ['check_recovery']

# считаю MAE
mae = mean_absolute_error(data_test_target, data_test_predicted)

print ('MAE равно:', round(mae, 4))
```

#### Вывод по проверке расчётов

MAE получилась довольно низкая, это значит, что данные посчитаны верно

### Суммарная концентраця всех веществ на разных стадиях

Нужно рассмотреть суммарную концентрацию всех веществ на столбцах (стадиях):

- rougher.input.feed_ - черновой концентрат до флотации в сырье
- rougher.output.concentrate_ - черновой концентрат после флотации
- final.output.concentrate_ - финальный концентрат после всех этапов

```python
# создаём списки для заполнения таблицы
col_list = ['rougher.input.feed_', 'rougher.output.concentrate_', 'final.output.concentrate_']
sub_list = ['au', 'ag', 'pb', 'sol']

# новая таблица для заполнения и наглядности
data_concent = pd.DataFrame()
data_concent['date'] = data_full['date']

# выбираем нужные столбцы
for col1 in col_list:
    for sub in sub_list:
        for col2 in data_full.columns:
            if (col1+sub) in col2:
                data_concent[col2] = data_full[col2]
            
# посчитаем общую концентрацию всех веществ
data_concent['rougher.input.feed'] = (
    data_concent['rougher.input.feed_au'] +
    data_concent['rougher.input.feed_ag'] +
    data_concent['rougher.input.feed_pb'] +
    data_concent['rougher.input.feed_sol']
)

data_concent['rougher.output.concentrate'] = (
    data_concent['rougher.output.concentrate_au'] +
    data_concent['rougher.output.concentrate_ag'] +
    data_concent['rougher.output.concentrate_pb'] +
    data_concent['rougher.output.concentrate_sol']
)

data_concent['final.output.concentrate'] = (
    data_concent['final.output.concentrate_au'] +
    data_concent['final.output.concentrate_ag'] +
    data_concent['final.output.concentrate_pb'] +
    data_concent['final.output.concentrate_sol']
)
```

### Напишите функцию для вычисления итоговой sMAPE.

```python
def sMAPE(target, predict):
    target = target.to_numpy()
    total = 0
    for x in range(len(target)):
        total += abs(target[x, 0] - predict[x, 0]) / (abs(target[x, 0]) + abs(predict[x, 0]) / 2)
    sMAPE_r = total / len(target) *100
    
    total = 0
    for x in range(len(target)):
        total += abs(target[x, 1] - predict[x, 1]) / (abs(target[x, 1]) + abs(predict[x, 1]) / 2)
    sMAPE_f = total / len(target) *100
    
    sMAPE_result = 0.25 * sMAPE_r + 0.75 * sMAPE_f
    return sMAPE_result
```
